```python
# II NLTK 
import nltk
texts = ["""John Christopher Depp II (born June 9, 1963) is an American actor, producer, and musician. He has been nominated for 10 Golden Globe Awards, winning one for Best Actor for his performance of the title role in Sweeney Todd: The Demon Barber of Fleet Street (2008) and has been nominated for three Academy Awards for Best Actor, among other accolades. He is regarded as one of the world's biggest film stars."""] #paste text after the three quotes, organise into lines if required
```

```python
for text in texts:
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        tagged_words = nltk.pos_tag(words)
        print(tagged_words)
```
