```python
from nltk.corpus import names
```

```python
def gender_features(word):
    return {'last_letter':word[-1]}
```

```python
gender_features('Obama')
```

```python
names.words('male.txt')
```

```python
labeled_names = ([(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
```

```python
featuresets = [(gender_features(name),gender) for (name,gender) in labeled_names]
featuresets[:3]
```

```python
train_set, test_set = featuresets[:5000], featuresets[5000:]
```

```python
import nltk
clf = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(clf,test_set))
```

```python
clf.show_most_informative_features(10)
```
